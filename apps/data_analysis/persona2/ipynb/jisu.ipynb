```python
import pandas as pd

# 첫 번째 파일 읽기
df_detail = pd.read_csv('./csv/질병분류코드별_진료내역_항목별_요양급여비용_현황_등록_후_12개월_2021.csv', header=1)

# 두 번째 파일 읽기
df_total = pd.read_csv('./csv/건강보험심사평가원_희귀질환 진료비 통계 2021.csv', encoding='CP949')
```

```python
df_total['희귀질환 상병코드'].rename('희귀질환_상병코드')
```

```python
df_total.query('희귀질환_상병코드 =="D66"')
```

```python
rare_code = df_total['희귀질환 상병코드'].unique().tolist()
rare_code
```

```python
for x in rare_code:
    # 조건에 맞는 인덱스를 찾고, 해당 인덱스에 대해 '희귀질환 상병코드' 열에 값을 할당합니다.
    df_detail.loc[df_detail['질병분류코드 및 질환별(2)'].str.contains(x), '희귀질환 상병코드'] = x
    df_detail.loc[df_detail['질병분류코드 및 질환별(2)'].str.contains(x), '환자수'] = x
```
