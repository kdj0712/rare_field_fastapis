- 연령별 발생자 비교
- 질병분류코드별 발생자 추이

##### 연령별 발생자 비교

```python
total_age01 = df_gender_age_accrual.loc[0][2:]
total_age01
```

```python
total_age02 = df_gender_age_accrual.loc[1][2:]
total_age02
```

```python
# 하나의 dataframe으로 만들어주기
df_total_age = pd.concat([total_age01, total_age02], axis=1)
df_total_age.columns = ['연령', '발생자수']
df_total_age
```

```python
df_total_age_cut= df_total_age[1:-1]
df_total_age_cut
```

```python
# 파이차트 그리기
plt.figure(figsize=(7, 7))
plt.pie(df_total_age_cut['발생자수'], labels=df_total_age_cut['연령'], autopct='%1.1f%%')
plt.title('2021 연령별 희귀질환 발생자')
plt.show()
```

```python
type(df_total_age_cut['발생자수'])
```

```python
plt.plot(df_total_age_cut['연령'], df_total_age_cut['발생자수'], marker='o', linestyle='-')
plt.title('2021 연령별 희귀질환 발생자')
plt.xlabel('연령')
plt.ylabel('발생자수')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(df_total_age_cut['연령'], df_total_age_cut['발생자수'], marker='o', linestyle='None')
plt.title('2021 연령별 희귀질환 발생자')
plt.xlabel('연령')
plt.ylabel('발생자수')
plt.xticks(rotation=45)
plt.show()
```
